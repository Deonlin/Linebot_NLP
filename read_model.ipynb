```python
from sklearn.externals import joblib
from bert_serving.client import BertClient
```

```python
clf2 = joblib.load('cls.pkl')
```

```python
def sentToVect(sent1,List=None):
    bc = BertClient()
    if List is None:
        List=[]
        List.append(sent1)
    sents_enc = bc.encode(List)
    return sents_enc
```

```python
sentToVect('健身房在哪裡')
```

```python
result = clf2.predict(sentToVect('健身房在哪裡'))
classes = ['房內服務行為','詢問健身房地點','詢問餐廳開放時間']
replyIntent = classes[result[0]]
print(replyIntent)
```
